# Exploratory

This notebook is an exploration of the doctor-nurse dataset. It first takes a look at the images from different categories, then computes some basic statistics about the dataset bias.

```python
%matplotlib inline
import os
import random
import shutil
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.figure as figure
from IPython.display import clear_output
import time
from IPython.display import display
import ipywidgets as widgets
```

### Defines the location pathes of the dataset 

```python
path_bask_r_f = 'basket/basket_f_r/'
path_bask_y_f = 'basket/basket_f_y/'
path_bask_r_m = 'basket/basket_m_r/'
path_bask_y_m = 'basket/basket_m_y/'

bask_r_f = os.listdir(path_bask_r_f)
bask_y_f = os.listdir(path_bask_y_f)
bask_r_m = os.listdir(path_bask_r_m)
bask_y_m = os.listdir(path_bask_y_m)

path_voll_r_f = 'volley/volley_f_r/'
path_voll_y_f = 'volley/volley_f_y/'
path_voll_r_m = 'volley/volley_m_r/'
path_voll_y_m = 'volley/volley_m_y/'

voll_r_f = os.listdir(path_voll_r_f)
voll_y_f = os.listdir(path_voll_y_f)
voll_r_m = os.listdir(path_voll_r_m)
voll_y_m = os.listdir(path_voll_y_m)
```

### Image display methods 

```python
def show_button(i, path, imgs):
    next_button = widgets.Button(description = 'Next')   
    next_button.on_click(lambda arg: clicked(i, path, imgs))
    display(next_button)


def clicked(i, path, imgs):
    i += 1
    img_show(i, path, imgs)
    show_button(i, path, imgs)


def img_show(i, path, imgs):
    clear_output(wait=True)
    image = Image.open(path + imgs[i], "r")
    plt.imshow(np.asarray(image))
    plt.show()
```

# Statistics 

###  Color as a protected attribute

### Basketball dataset

```python
fig, ax = plt.subplots()
bask_r, bask_y = len(bask_r_f) + len(bask_r_m), len(bask_y_f) + len(bask_y_m)
ax.bar([0,1], [bask_r, bask_y], tick_label=["Red", "Yellow"])
ax.set_title("Basketball dataset")
print("Proportions: {:.2f}% red shirt and {:.2f}% yellow shirt.".format(bask_r/(bask_r + bask_y)*100, bask_y/(bask_r + bask_y)*100))
```

### Volleyball dataset

```python
fig, ax = plt.subplots()
voll_r, voll_y = len(voll_r_f) + len(voll_r_m), len(voll_y_f) + len(voll_y_m)
ax.bar([0,1], [voll_r, voll_y], tick_label=["Red", "Yellow"])
ax.set_title("volletball dataset")
print("Proportions: {:.2f}% red shirt and {:.2f}% yellow shirt.".format(voll_r/(voll_r + voll_y)*100, voll_y/(voll_r + voll_y)*100))
```

###  Gender as a protected attribute

### Basketball dataset

```python
fig, ax = plt.subplots()
bask_f, bask_m = len(bask_r_f) + len(bask_y_f), len(bask_r_m) + len(bask_y_m)
ax.bar([0,1], [bask_f, bask_m], tick_label=["Female", "Male"])
ax.set_title("Basketball dataset")
print("Proportions: {:.2f}% females and {:.2f}% males.".format(bask_f/(bask_f + bask_m)*100, bask_m/(bask_f + bask_m)*100))
```
